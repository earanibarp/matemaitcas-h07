# Alquileres

Este dataset contiene información sobre el valor, mantenimiento e impuestos de los distintos tipos de lugares, como habitaciones, casas, locales comerciales que se encuentran en los distritos de Lima

Contiene las siguientes columnas:


*   Tipo
*   Distrito
*   Cuartos
*   Vacantes
*   Suites
*   Area
*   Valor
*   Mantenimiento
*   Impuesto



```python
# Importación de las librerías a utilizar
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Carga de datos
df = pd.read_csv("alquiler.csv", sep=";")
# Exploración de datos
df.head()
```

```python
# Exploración de datos
df.info()
```

```python
df.isnull().sum()
```

```python
# Llenado de datos faltantes
df["Valor"].fillna(0, inplace = True)
df["Mantenimiento"].fillna(0, inplace = True)
df["Impuesto"].fillna(0, inplace = True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df2 = pd.read_csv("alquiler.csv", sep=";")
df2.info()
```

```python
df2.isnull().sum()
```

```python
# Eliminado de datos faltantes
df2.dropna(subset=["Valor", "Mantenimiento", "Impuesto"], inplace = True)
```

```python
df2.shape
```

```python
# Verificar los resultados
df.head()
```

```python
# Agrupar datos
df.groupby(by=["Tipo", "Distrito"] ).sum()
```

```python
cuartos = df.groupby(by="Distrito" ).Cuartos.sum()
plt.figure(figsize=(70,10))
plt.bar(cuartos.index, cuartos)
plt.show()
```

```python
df["Tipo"].drop_duplicates().values
```

```python
residencial = ['Habitación', 'Casa', 'Departamento', 'Casa en condominio', 'Casa de villa', 'Loft']
```

```python
select = df["Tipo"].isin(residencial)
df_residenciales = df[select]
df_residenciales
```

```python
df_residenciales.groupby(by=["Tipo", "Distrito"]).sum().plot(figsize=(30,10))
```

```python
mantenimiento = df_residenciales["Mantenimiento"] > 0
df_residenciales[mantenimiento].groupby("Distrito").Mantenimiento.sum().plot.barh(figsize=(5,10))
plt.title("Mantenimiento de residenciales de los distritos de Lima")
plt.xlabel("Total mantenimiento")
plt.show()
```

```python
df_residenciales["Cuartos"].plot.hist(bins=10)
plt.xlabel("# de Cuartos")
plt.ylabel("Frecuencia")
plt.title("Distribución de cuartos")
plt.show()
```

```python
df_residenciales.groupby("Tipo").Cuartos.count().plot.bar()
```

```python
df_residenciales[["Valor", "Area"]].plot.scatter("Area", "Valor")
```

```python
plt.figure(figsize=(5,10))
plt.scatter(df_residenciales["Area"], df_residenciales["Valor"])
plt.title("Relación de Valor vs Area")
plt.xlabel("Area (metros cuadrados)")
plt.ylabel("Valor en S/.")
plt.show()
```
