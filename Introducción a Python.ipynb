````python
"""
efectivo           -> 10% dscto
tarjeta de débito  -> 10% dscto
tarjeta de crédito -> 8% dscto
billetera digital  -> 5% dscto
"""

medioPago = input("Ingresa el medio de pago: ")

if medioPago == "EFECTIVO":
  print("Usted tiene 10% de descuento")
elif medioPago == "TARJETA DE DEBITO":
  print("Usted tiene 10% de descuento")
elif medioPago == "TARJETA DE CREDITO":
  print("Usted tiene 8% de descuento")
else:
  print("Usted tiene 5% de descuento")

````

````python
# @title Estructura cíclica: FOR

for i in range(4): # range(inicio, final, paso)
  print(f"Hola mundo, {i}")

for i in [0, 10, 20, 30]:
  print(i)
````

````python
# @title Estructura cíclica: WHILE

x = 5

while x >= 0:
  print("Hola mundo")
  x-- # x = x - 1
  if x == 3:
     break
````

````python
# @title Listas

lista = [1, 3, 5, 7, 'gato']
lista.append(2) # insertar un elemento al final de la lista
lista.insert(0, "perro") #inserta un elemento en una posición indicada

lista.pop(-1) # eliminar un elemento teniendo en cuenta su índice

lista.remove("gato") # eliminar un elemento que se encuentre en la lista

lista.reverse()
lista.clear() # limpiar la lista
lista
````

````python
productos = {
    "nombre": ["lapiz", "lapicero", "borrador"],
    "precio": [1,         2,          0.5],
    "cantidad": [10,      5,          15]
}

nombreP = input("Ingrese el producto: ")

# productos["nombre"][2]

if nombreP in productos["nombre"]:
  cantidad_venta = int(input("Ingrese la cantidad: ")) # 2
  indice = productos["nombre"].index(nombreP) # indice = 1
  calculo = productos["cantidad"][indice] - cantidad_venta # calculo = 5 - 2
  productos["cantidad"][indice] = calculo # 5 = 3
  print(productos)
else:
  print("El producto no existe")



````
